[4] DATA PREPROCESSING

```python
data["Character"] = data["Character"].str.strip().str.capitalize()
data["Interaction With"] = data["Interaction With"].str.strip().str.capitalize()
data
```

```python
sentiment_mapping = {"Positive": 1, "Negative": -1, "Neutral": 0}
data["Sentiment Score"] = data["Sentiment"].map(sentiment_mapping)
data
```

```python
data['Sentiment_Category'] = pd.cut(data['Sentiment Score'], bins=[-2,-1, 0, 1], labels=['Low', 'Medium', 'High'])
data
```
